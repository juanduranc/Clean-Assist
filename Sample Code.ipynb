```python
import seaborn as sns
mpg = sns.load_dataset('mpg')
mpg.head(5)
```

<span style="color:darkblue; font-size:14px"><b>Step #2:</b></span><br>
<span style="font-size:14px">1) Open jupyter Notebook and run: pip install broompy</span><br>
<span style="font-size:14px">2) Import broompy library.</span>

```python
import broompy
help(broompy.core.table)
help(broompy.core.normality)
help(broompy.core.scatter)
```

<span style="color:darkblue; font-size:14px"><b>Step #3: Implementation of functions</b></span><br><br>
<span style="color:green; font-size:14px"><b>A) Implementation of broompy table</b></span><br>
<span style="font-size:14px">broompy.core.table</span>

```python
dataframe = mpg
broompy.core.table(dataframe, 10, 4)
```

<span style="color:green; font-size:14px"><b>B) Implementation of broompy normality</b></span><br>
<span style="font-size:14px">broompy.core.table</span>

```python
list_var = ['displacement', 'cylinders', 'model_year', 'weight']
broompy.core.normality(dataframe, list_var, "n", 10, 30, 10)
```

<span style="color:green; font-size:14px"><b>C) Implementation of broompy normality</b></span><br>
<span style="font-size:14px">broompy.core.scatter</span>

```python
data_frame = mpg
print_img = 'n'
size_x = 10
size_y = 16
size_font = 14
color = 'b'
response_var = 'displacement'

explanatory_var = ['cylinders', 'model_year']
broompy.core.scatter(data_frame, explanatory_var, response_var, print_img, size_x, size_y, size_font, 'g')

explanatory_var = ['weight', 'acceleration']
broompy.core.scatter(data_frame, explanatory_var, response_var, print_img, size_x, size_y, size_font, 'b')
```
